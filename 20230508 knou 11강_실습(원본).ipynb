math 모듈을 사용한 원뿔 계산 프로그램

```python
# 원뿔 클래스 정의
class Cone :
    def __init__(self, radius = 20, height = 30):
        self.r = radius
        self.h = height
    
    def get_vol(self) :
        return 1/3 * math.pi * self.r ** 2 * self.h

    def get_surf(self) :
        return math.pi * self.r ** 2 + math.pi * self.r * self.h
```

```python
class Triangle :
    def __init__(self, a = 1, b = 1, alpha = 60) :
        if a > 0 and b > 0 :
            self.__a = a
            self.__b = b
        else :
            print("양수로 a, b를 입력해 주세요")
        
        if 0 < alpha and alpha < 180 :
            self.__alpha = alpha
        else :
            print("범위에 맞게 각도를 입력해 주셔야 합니다")

    def get_a (self) :
        return self.__a

    def get_b (self) :
        return self.__b

    def get_alpha (self) :
        return self.__alpha

    def set_a (self, a) :
        if a > 0 :
           self.__a = a

    def set_b (self, b) :
        if b > 0 :
           self.__b = __b
    
    def set_alpha (self, alpha) :
        if 0 < alpha and alpha < 180 :
            self.__alpha = alpha
    
    def get_surf(self) :
        return 1 / 2 * self.__a * self.__b * math.sin(self.__alpha)
```

```python
a, b = input("삼각형의 두 변의 길이를 입력해 주세요 > ").split()

# references
# taking two inputs at a time (https://www.geeksforgeeks.org/taking-multiple-inputs-from-user-in-python/)
# a, b = input("Enter two values: ").split()
# print("First number is {} and second number is {}".format(a, b))

alpha = int(input("0보다 크고 180보다 작은 각도 하나를 정수로 입력해 주세요 > "))
myTriangle = Triangle(int(a), int(b), alpha)
print("myTriangle의 두 변의 길이는 ", myTriangle.get_a(), "와 ", myTriangle.get_b(), "이고, 그 끼인각은 ", myTriangle.get_alpha(), "이며, 이 때 이 삼각형의 넓이는 ", myTriangle.get_surf(), "입니다", sep="")
```

```python
from random import *
```

```python
# 사용자로부터 숫자 입력받기
my_num_list = list(map(int, input("1~45 범위의 정수 6개를 입력해 주세요(띄어쓰기로 구분해 주세요) > ").split())) # 쉼표로 구분받고 싶으면 split(",")
# print(my_num_list) # e.g. [42, 1, 6, 22, 30, 31]

# references
# 값 두 개를 입력받아 변수 a와 b에 저장 (띄어쓰기 구분) (https://happyeuni.tistory.com/18)
# a, b = input().split() #문자열로
# a, b = map(int, input().split()) # 정수형으로
# 1차원 배열 입력받기 = 정수형 리스트로 저장
# num_list = list(map(int, input().split())) # 입력 : 1 2 3 /출력 : [1, 2, 3] 

# 로또 당첨 번호 생성/추첨
lotto_num_list = []
for i in range (0, 6) :
    lotto_num_list.append(randint(1, 45))
print("당첨 숫자는 " + str(lotto_num_list) + "입니다")

# 사용자의 맞은 개수 구하기
# 방법1
count1 = 0
for my_num in my_num_list :
    for lotto_num in lotto_num_list :
        if my_num == lotto_num :
            count1 = count1 + 1
print(str(count1) + "개가 맞았습니다!")

# 방법2
count2 = 0
for my_num in my_num_list :
    if my_num in lotto_num_list :
        count2 = count2 + 1
print(str(count2) + "개가 맞았습니다!") 
```

```python
answer = randint(1, 1000)
print("answer = ", answer, sep="")

for i in range (1, 21) :
    guess = int(input("숫자를 맞혀보세요(1~1000 사이의 정수, %d번째 시도) > " % i))
    if guess > answer :
        print("%d보다 작습니다. " % guess, end="")
    elif guess < answer :
        print("%d보다 큽니다. " % guess, end="")
    else :
        print("맞았습니다. 정답은 %d입니다! %d번만에 맞추셨네요, great job!" % (guess, i))
        break
```

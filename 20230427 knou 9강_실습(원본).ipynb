**세 개의 사용자 입력을 오름차순으로 정렬하는 함수를 이용하여 정렬된 값을 출력하는 문제**


### 15h10 v1 은영 실습

```python
def sort3_v1(a, b, c) :
    if (a < b) : # 파이썬에서는 조건절에 괄호 필요 없음
        if (b < c) :
            size1, size2, size3 = a, b, c
        else :
            if (a < c) :
                size1, size2, size3 = a, c, b
            else :
                size1, size2, size3 = c, a, b
    else :
        if (b > c) :
            size1, size2, size3 = c, b, a
        else :
            if (a < c) :
                size1, size2, size3 = b, a, c
            else :
                size1, size2, size3 = b, c, a

    return size1, size2, size3
```

### 15h35 수업 들은 후 v2 은영 실습
파이썬의 동시할당 개념을 활용하여 보다 더 간단한 방법으로 풀어볼 수 있음

```python
def sort3_v2(a, b, c) :
    if a > b :
        a, b = b, a
    if a > c :
        a, c = c, a
    if b > c :
        b, c = c, b

    print("v2 정렬 결과 =", a, b, c)
```

```python
a = int(input("첫번째 숫자를 입력하세요: "))
b = int(input("두번째 숫자를 입력하세요: "))
c = int(input("세번째 숫자를 입력하세요: "))

print("v1 정렬 결과 =", sort3_v1(a, b, c)) # v1 정렬 결과 = (5, 7, 10)
sort3_v2(a, b, c) # v2 정렬 결과 = 5 7 10

print("입력했던 숫자들 =", a, b, c) # 입력했던 숫자들 = 10 5 7 <- 함수 내부에는 값만 전달되는 바, 변수에는 영향 가지 않음
```

### 16h55 v1 은영 실습

```python
# 원뿔 계산 함수 정의
def rtn_cone_vol_surf(r = 20, h = 30) :
    if r > 0 and h > 0 :
        # r, h 모두 양수일 때
        return 1/3 * 3.14 * r ** 2 * h, 3.14 * r ** 2 + 3.14 * r * h
    else :
        # r, h가 음수일 때
        print("반지름과 높이 값에 양수를 입력하세요")

print(rtn_cone_vol_surf(100, 200)) # (2093333.3333333333, 94200.0)
print(rtn_cone_vol_surf()) # (12559.999999999998, 3140.0)
```

**여러개의 수를 입력받고 합과 평균을 반환하는 var_sum_avg 함수를 사용하여, (20, 25, 10, 85, 100, 150)에 대한 합과 평균을 출력하는 프로그램을 작성하시오.**


### 17h v1 은영 실습

```python
def var_sum_avg(*nums) :
    sum = 0
    for num in nums :
        sum += num
    avg = sum / len(nums) # 교수님께서는 count 변수 하나 만드셔서 for문 돌 때마다 계수하심
    return sum, avg
```

```python
sum, avg = var_sum_avg(20, 25, 10, 85, 100, 150)
print("합 = ", sum, ", 평균 = ", avg, sep="")
```

```python
# 파라미터의 개수가 정해지지 않았다면 = 파라미터의 개수가 가변적인 형태의 함수 -> 파라미터로 넘어오는 값들은 시퀀스 형태로 전달됨 + 반복 사용됨
def var_sum_avg(*numbers) :
    sum = 0
    count = 0

    for i in numbers :
        sum = sum + i
        count = count + 1

    return sum, sum/count

print(var_sum_avg(10, 20, 30, 40)) # (100, 25.0)
```

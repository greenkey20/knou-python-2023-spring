### 내가 생각한 로직

1. 텍스트를 모두 소문자로 바꿈 <- .lowercase()
2. 정규표현식 re.findall(r'\w+', str) 사용해서 영어 단어들을 골라내기
3. 아래 로직 수행

```python
for word in text: # 텍스트의 모든 단어를 하나한 순회하며
    if word in dictionary: # 그 단어가 나의 dictionary에 있으면
        dictionary.get(word) += 1 # 해당 word를 key로 갖는 항목의 value 값을 1 증가시킴
    else: # 그 단어가 나의 dictionary에 아직 없으면
        dictionary(word) = 1 # 해당 word를 key로 갖는 항목을 추가 + 초기 value 값은 1
```

````python
import re
````

````python
h_fp = open("Hamlet_by_Shakespeare.txt", "r")

h_dict = dict() # Hamlet에 등장하는 단어 및 그 횟수를 관리하는 딕셔너리

for line in h_fp.readlines(): # 파일 내용을 1줄씩 읽어서 line에 저장
    words = re.findall(r'\w+', line.lower()) # line을 소문자로 바꾸고, 해당 정규표현식 적용 -> 패턴과 일치하는 + 소문자로 된 단어들의 list인 'words'가 나옴

    for word in words:
        if word in h_dict: # words 리스트에 있는 단어 word가 h_dict에 존재하면
            cnt = h_dict.get(word) # 현재 저장된 cnt 값을 꺼냄
            h_dict[word] = cnt + 1
        else: # words 리스트에 있는 단어 word가 h_dict에 존재하지 않으면
            h_dict[word] = 1 # 해당 단어의 cnt를 1로 초기화해서 딕셔너리 항목으로 추가

for key in h_dict: # Hamlet 딕셔너리에 있는 key-value 쌍 내용들 하나하나 꺼내어 출력
    print("[" + key + "] " + str(h_dict.get(key)) + "회")


h_fp.close()
````

'hamlet_by_Shakespeare.txt' 파일에 포함된 단어 중 출현한 횟수가 100 이상되는 단어와 출력 횟수를 출력하는 프로그램을 작성하시오 


````python
freq_h_dict = dict()

for key in h_dict:
    cnt = h_dict.get(key)

    if cnt >= 100:
        freq_h_dict[key] = cnt

sorted_h_dict = sorted(freq_h_dict.items(), key=lambda x: x[1], reverse=True)
print(sorted_h_dict) # 이렇게 찍어보니 교수님께서 제시해주신 예시와 cnt가 다르다 + list 항목을 딕셔너리 항목(key:value)으로 바꾸는 방법 교수님께서 알려주신 것 사용해보자

# print("[" + key + "] " + str(h_dict.get(key)) + "회")
````
